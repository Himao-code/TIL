- 데이터 정제하기
    - 불필요한 데이터 삭제

```python
import pandas as pd
df = pd.read_csv('./namsan.csv', encoding='EUC-KR', low_memory=False)
```


## 데이터 삭제

```python
df.head(3)
```

```python
# 슬라이싱으로 가능은 함. 
df.loc[:, '번호':'등록일자'].head()
```

```python
# 컬럼들
print(df.columns)
mask = df.columns != 'Unnamed: 13'
print(mask)
df = df.loc[:, mask]
df.head(3)
```

```python
# 진짜로 컬럼 삭제
df.drop(['부가기호', 'Unnamed: 13'], axis=1)

# 리턴말고 원본에서 삭제하기
df.drop(['부가기호', 'Unnamed: 13', '주제분류번호'], axis=1, inplace=True)
```

```python
df.head(3)
```

```python
# col중에 하나라도 NaN(비어있음)이면 그 Col 삭제
book_df = df.dropna(axis=1)
book_df.head(3)
# col의 모든 데이터가 NaN이면 해당 col 삭제
book_df = df.dropna(axis=1, how='all')
book_df.head(3)
```
